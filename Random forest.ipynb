```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
#creating pandas dataframe from this data
df = pd.DataFrame(digits.data)
df.head()
```

```python
#creating target value in dataframe
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis = 'columns'),digits.target,test_size = 0.2)
```

```python
len(X_test)
```

```python
# using random forest classifier to train model
#ensemble used when we use multiple algorithm to predict outcome
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
# getting 96% accuracy
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
# we give a truth and then prediction
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import seaborn as sn 
plt.figure(figsize=(10,9))
sn.heatmap(cm, annot = True)
plt.xlabel('Predict')
plt.ylabel('Truth')
```
